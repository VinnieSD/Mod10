````python
import pandas as pd
from scipy.stats import ttest_ind

df = pd.read_csv("previsao_de_renda.csv")

grupo_1 = df[df['posse_de_imovel'] == True]['renda']
grupo_2 = df[df['posse_de_imovel'] == False]['renda']

teste_t_posse_imovel = ttest_ind(grupo_1, grupo_2, equal_var=False)

print("Resultado do Teste t para 'posse_de_imovel':")
print(teste_t_posse_imovel)

grupo_3 = df[df['posse_de_veiculo'] == True]['renda']
grupo_4 = df[df['posse_de_veiculo'] == False]['renda']

teste_t_posse_veiculo = ttest_ind(grupo_3, grupo_4, equal_var=False)
print("\nResultado do Teste t para 'posse_de_veiculo':")
print(teste_t_posse_veiculo)
````
