Considere a mesma base de dados utilizada na tarefa 1 ```previsao_de_renda.csv```.

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("previsao_de_renda.csv")

data_grouped = data.groupby(['data_ref', 'posse_de_imovel', 'posse_de_veiculo']).size().unstack().fillna(0)

data_grouped.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribuição das Variáveis Qualitativas ao Longo do Tempo')
plt.xlabel('Data de Referência')
plt.ylabel('Contagem')
plt.legend(title='Legenda')
plt.show()
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

Resposta: Em posse de veículos ambos desviam em na mesma direcao, na maioria das vezes.
Em relacao aos imóveis, dificilmente os desvios sao muito similares, apesar de haverem alguns momento em que seguem a mesma direcao.

````python
imovel = data.groupby(['data_ref', 'posse_de_imovel'])['renda'].mean().unstack().fillna(0)
veiculo = data.groupby(['data_ref', 'posse_de_veiculo'])['renda'].mean().unstack().fillna(0)
````

````python
imovel.plot(kind='line', marker='o', figsize=(12, 5))
plt.title('Médias de Renda por Posse de Imóvel ao Longo do Tempo')
plt.xlabel('Data de Referência')
plt.ylabel('Média de Renda')
plt.legend(title='Posse de Imóvel')

veiculo.plot(kind='line', marker='o', figsize=(12, 5))
plt.title('Médias de Renda por Posse de Veículo ao Longo do Tempo')
plt.xlabel('Data de Referência')
plt.ylabel('Média de Renda')
plt.legend(title='Posse de Veículo')

plt.tight_layout()
plt.show()
````
